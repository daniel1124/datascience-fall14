```python
# Import linear regression and the diabetes dataset
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes

# Load the diabetes dataset
diabetes = load_diabetes()

# Use only one feature -- the following code creates a 1-dimensional 
# array containing just the second feature
diabetes_X = diabetes.data[:, np.newaxis]
diabetes_X_temp = diabetes_X[:,:,2]

# Split the data into training/testing sets -- we will use all but last 20 elements for training
diabetes_X_train = diabetes_X_temp[:-20]
diabetes_X_test = diabetes_X_temp[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]

# Create linear regression object
regr = LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)
```

```python
# The coefficients
print 'Coefficients:', regr.coef_
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((regr.predict(diabetes_X_test) - diabetes_y_test) ** 2))
```

```python
# Predict the values of unseen items
predicted_targets = regr.predict(diabetes_X_test)

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')
plt.plot(diabetes_X_test, predicted_targets , color='blue',
         linewidth=3)

plt.show()
```

```python
# Split the data into training/testing sets
diabetes_X_train = diabetes.data[:-20]
diabetes_X_test = diabetes.data[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]

# Create linear regression object
regr = LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)
```

```python
# The coefficients
print 'Coefficients:', regr.coef_
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((regr.predict(diabetes_X_test) - diabetes_y_test) ** 2))
```

On the training data, including more features $R^2$ will definitely increase (RSS decrease). This is not necessarily true on the testing data. However here our testing data is fairly small, which is not enough for any justification.  

```python
from sklearn.feature_selection import RFE

# Create the RFE object using two features and rank the features for the diabetes datasets
rfe = RFE(estimator=regr, n_features_to_select=2, step=1)
rfe.fit(diabetes_X_train,diabetes_y_train)
print rfe.ranking_
```

```python
rank = rfe.ranking_
rank_dict = dict()

for i in range(10):
    if rank[i] not in rank_dict.keys():
        rank_dict[rank[i]] = [i]
    else:
        rank_dict[rank[i]].append(i)

def flatten(l):
    flattenL = list()
    for subl in l:
        if type(subl) is list:
            for item in subl:
                flattenL.append(item)
        else:
            flattenL.append(subl)
    return flattenL


selset = list()
rss = list()
feat_num = list()
for i in range(len(rank_dict)):
    selset.append(list(rank_dict[i+1]))
    selset = flatten(selset)
    
    diabetes_X_temp = diabetes.data[:,selset]
    diabetes_X_train = diabetes_X_temp[:-20]
    diabetes_X_test = diabetes_X_temp[-20:]
    diabetes_y_train = diabetes.target[:-20]
    diabetes_y_test = diabetes.target[-20:]
    regr.fit(diabetes_X_train, diabetes_y_train)
    feat_num.append(len(selset))
    print "Number of Features: ", feat_num[-1]
    print 'Coefficients:', regr.coef_
    rss.append(np.mean((regr.predict(diabetes_X_test) - diabetes_y_test) ** 2))
    print("Residual sum of squares: %.2f"
      % rss[-1])
    print '\n'

plt.plot(feat_num,rss)
plt.xlabel('Number of Features')
plt.ylabel("RSS")
```

From the results we see using 2 features has lower RSS than using 10 features. The best one is using three features which has the lowest RSS.

## Iris

```python
sp = 0.2
iris = load_iris()
data = np.concatenate((iris.data,iris.target[:,np.newaxis]),axis=1)
data = np.random.permutation(data)
iris.data = data[:,0:4]
iris.target = data[:,4]

nrow = iris.data.shape[0]
iris_X_train = iris.data[:-sp*nrow,:]
iris_X_test = iris.data[-sp*nrow:,:]
iris_y_train = iris.target[:-sp*nrow]
iris_y_test = iris.target[-sp*nrow:]


svm_linear = svm.SVC(kernel = "linear")
svm_linear.fit(iris_X_train,iris_y_train)
pred = svm_linear.predict(iris_X_test)

def cal_error(test_y, pred):    
    errors = 0.0
    for i in range(len(pred)):
        if pred[i] != test_y[i]:
            errors += 1.0
    return errors

print "Linear SVM"
errors = cal_error(iris_y_test,pred)
print "Test data errors = ",errors, "out of", len(pred), "predictions"

total_pred = svm_linear.predict(iris.data)
errors = cal_error(iris.target,total_pred)
print "Total errors = ",errors, "out of", len(total_pred), "predictions"


svm_RBF = svm.SVC(kernel = "rbf")
svm_RBF.fit(iris_X_train,iris_y_train)

pred = svm_RBF.predict(iris_X_test)
svm_rbf_pred = svm_RBF.predict(iris.data)

print "Non-linear SVM"
errors = cal_error(iris_y_test,pred)
print "Test data errors = ",errors, "out of", len(pred), "predictions"
errors = cal_error(iris.target,total_pred)
print "Total errors = ",errors, "out of", len(svm_rbf_pred), "predictions"
```

## Decision Tree

```python
from sklearn.datasets import make_classification

# Generate synthetic dataset
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, random_state=5, n_classes=2)

# Plot synthetic dataset
plt.scatter(X[:,0],X[:,1],marker='o',c=colormap[y], s=50)
plt.show()
```

```python
# Import kmeans algorithm and clustering evaluation metrics
from sklearn.cluster import KMeans
from sklearn import metrics

kmeans = KMeans(init='random', n_clusters = 2, n_init=10)
kmeans.fit(X)

# Evaluate clusters: see http://scikit-learn.org/stable/modules/clustering.html#homogeneity-completeness-and-v-measure for more details

# Homogeneity score: evaluates if each cluster contains only memebers of a single class
print "Homogeneity_score = ",metrics.homogeneity_score(y,kmeans.labels_)

# Completeness score: evaluates if all members of a given class are assigned to the same cluster
print "Completeness_score = ",metrics.completeness_score(y,kmeans.labels_)
```

```python
# Plot the decision boundary. For that, we will assign a color to each
h = .02
x_min, x_max = -2, 4
y_min, y_max = -2, 4
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')

plt.scatter(X[:,0],X[:,1],marker='o',c=colormap[y], s=50)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='w', zorder=10)

plt.title('K-means clustering on synthetic data\n'
          'Centroids are marked with white cross')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```
