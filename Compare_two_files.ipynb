```python
# Compare two files of stock positions, and report the changes of positions.
# The company names are ignored for convenience.
# Assume the file are well formated so that all lines has the same width.
def compare_two_files(path1,path2,k,w):
    # path1 and path2 specify file paths for the two documents
    # k: the number of lines read each time
    # w: line width
    
    report = []
    
    with open(path1) as f1, open(path2) as f2:
        
        # read k lines
        a1 = f1.read(w*k) 
        a2 = f2.read(w*k)
    
        while a1 and a2:

            if hash(a1) != hash(a2): # Fortunately this does not happen often
                
                # convert buffer to list
                l1 = [map(int,item.split()) for item in a1.split('\n')[:-1] ]
                l2 = [map(int,item.split()) for item in a2.split('\n')[:-1] ]
                
                # move the file pointers so ids can be matched again
                a1 = map(int, f1.read(w).split())
                a2 = map(int, f2.read(w).split())                       
                while a1 and a2 and a1[0] != a2[0]:
                    if a1[0] < a2[0]:
                        l1.append(a1)
                        a1 = map(int,f1.read(w).split())
                    elif a1[0] > a2[0]:
                        l2.append(a2)
                        a2 = map(int,f2.read(w).split())
                if a1:
                    l1.append(a1)
                if a2:
                    l2.append(a2)
                
                # find differences and send them to report
                p1,p2 = 0,0
                n1,n2 = len(l1),len(l2)
                while p1 < n1 and p2 < n2:
                    if l1[p1][0] == l2[p2][0]:
                        if l1[p1][1] != l2[p1][1]:
                            report.append( [ l1[p1][0], l1[p1][1], l2[p2][1]] )
                        p1,p2 = p1+1,p2+1
                    elif l1[p1][0] < l2[p2][0]:
                        report.append( [ l1[p1][0], l1[p1][1], 0 ] )
                        p1 += 1
                    else:
                        report.append( [ l2[p2][0], 0, l2[p2][1] ] )
                        p2 += 1
    
                while p1 < n1:
                    report.append( [ l1[p1][0], l1[p1][1], 0 ] )
                    p1 += 1
                
                while p2 < n2:
                    report.append( [ l2[p2][0], 0, l2[p2][1] ] )
                    p2 += 1
                
            a1 = f1.read(w*k)
            a2 = f2.read(w*k)

        # after the while loop, there might be some reminders in either file        
        if a1:
            a1 += f1.read()
            l1 = [map(int,item.split())+[0] for item in a1.split('\n')[:-1] ]
            report += l1
        if a2:
            a2 += f2.read()
            l2 = [[int(item.split()[0])]+[0]+[int(item.split()[1])] for item in a2.split('\n')[:-1] ]        
            report += l2
        
        return report
```

```python
# Test 
import numpy as np
path1 = 'stock_positions1.txt'
path2 = 'stock_positions2.txt'

with open(path1) as f1, open(path2) as f2:
    a1 = f1.read()
    a2 = f2.read()

print 'Stock Position 1\n',a1
print 'Stock Position 2\n',a2

k = 3
w = 13
print 'Change of Positions\n',['id','Position 1', 'Position 2'],'\n',np.array(compare_two_files(path1,path2,k,w))
```
